## Reconstruct database

```julia
name_lost_data = "Turning_rate_higher_low_noise_01_250000_time_seed_8436"
```

```julia
 lost_data_path  =  data_path *name_lost_data
```

```julia
settings = s.deserialize(lost_data_path * "\\settings.jls")
settings2 = deepcopy(settings)
```

```julia
settings2["path"] = lost_data_path
settings2["date"] = unix2datetime(time())
settings2["unix_date"] = time() 
settings2["run_time"] = 72000
settings2["storage"] = foldersize(lost_data_path)
settings2["location"] = data_path
settings2["name"] = name_lost_data[1:end-15]
```

```julia
push!(data_base,settings2, promote = true, cols = :union )
```

```julia

s.serialize("data_base/df.jls",data_base)
CSV.write("data_base/df.csv",data_base)
```
